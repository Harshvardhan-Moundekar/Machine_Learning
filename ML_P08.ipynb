```python
#AI-Driven Climate Prediction Models
#libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

```

```python
file_path = "/content/drive/MyDrive/6TH SEM/Machine Learning Lab/climate_change_dataset.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# 1. Handle Missing Data (if any)
# Check for missing values in each column
print("Missing values before preprocessing:")
print(df.isnull().sum())

#There is no requirement on handling missing values
```

```python
# 2. Encoding Categorical Data
# Use OneHotEncoding for the 'Country' column to convert categorical values into numerical
encoder = OneHotEncoder(drop='first', handle_unknown='ignore') # sparse=False is removed
country_encoded = encoder.fit_transform(df[['Country']]).toarray() # add .toarray() to get a dense array
country_df = pd.DataFrame(country_encoded, columns=encoder.get_feature_names_out(['Country']))
```

```python
#Feature Scaling (Standardize numerical features)
scalar = StandardScaler()
scaled_features = scalar.fit_transform(df[['Avg Temperature (°C)', 'CO2 Emissions (Tons/Capita)', 'Sea Level Rise (mm)', 'Rainfall (mm)', 'Population', 'Renewable Energy (%)', 'Extreme Weather Events', 'Forest Area (%)']])
scaled_df = pd.DataFrame(scaled_features, columns = ['Avg Temperature (°C)', 'CO2 Emissions (Tons/Capita)', 'Sea Level Rise (mm)', 'Rainfall (mm)', 'Population', 'Renewable Energy (%)', 'Extreme Weather Events', 'Forest Area (%)'])
df_preprocessed = pd.concat([scaled_df, country_df, df[['Year', 'CO2 Diff', 'Temp Diff', 'Rainfall Diff']]], axis=1)


print(df_preprocessed.head())

```
