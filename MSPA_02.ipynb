Implement Ensemble methods (bagging and boosting)

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file_path = "/content/drive/MyDrive/6TH SEM/Machine Learning Lab/train_clean_data.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
x = df.drop(['Loan_status_Y','Loan_ID','Dependents'],axis=1)
y = df['Loan_status_Y']

#split dataset into train and test(70 - 30)
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y, test_size=0.3, random_state=42)
```

```python
#Applying Random Forest Algorithm
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(x_train, y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix

print(confusion_matrix(y_test,rfc_pred))
```

```python
#Accuracy of the model
from sklearn.metrics import accuracy_score
print("Accuracy of Random Forest Algorithm : ",accuracy_score(y_test,rfc_pred))
```

```python
#Applying Adaboost Algorithm
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(x_train, y_train)
```

```python
ada_pred = ada.predict(x_test)
print(confusion_matrix(y_test,ada_pred))
```

```python
print("Accuracy of AdaBoost Algorithm : ", accuracy_score(y_test,ada_pred))
```

```python
#Apply XGBoost Algorithm
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=100)
xgb.fit(x_train, y_train)

```

```python
xgb_pred = xgb.predict(x_test)
print(confusion_matrix(y_test,xgb_pred))
```

```python
print("Accuracy of XGBoost Algorithm : ", accuracy_score(y_test,xgb_pred))
```

```python
#Apply Bagging and Boosting techniques
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier

#Apply those model
bag = BaggingClassifier(n_estimators=100)
bag.fit(x_train, y_train)
```

```python
gbc = GradientBoostingClassifier(n_estimators=100)
gbc.fit(x_train, y_train)
```

```python
bag_pred = bag.predict(x_test)
gbc_pred = gbc.predict(x_test)

print(confusion_matrix(y_test,bag_pred))
print(confusion_matrix(y_test,gbc_pred))

print("Accuracy of Bagging Algorithm : ", accuracy_score(y_test,bag_pred))
print("Accuracy of Gradient Boosting Algorithm : ", accuracy_score(y_test,gbc_pred))
```

```python
#Give me a Accuracy of all model
print("Accuracy of Random Forest Algorithm : ",accuracy_score(y_test,rfc_pred))
print("Accuracy of AdaBoost Algorithm : ", accuracy_score(y_test,ada_pred))
print("Accuracy of XGBoost Algorithm : ", accuracy_score(y_test,xgb_pred))
print("Accuracy of Bagging Algorithm : ", accuracy_score(y_test,bag_pred))
print("Accuracy of Gradient Boosting Algorithm : ", accuracy_score(y_test,gbc_pred))
```
